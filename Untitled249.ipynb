````python
#!/usr/bin/env python3
"""
AI Training Tools for Contradiction Detection

Provides:
- Training-example extraction from MBT recursive learning logs
- Stable JSON/CSV/TSV/OpenAI/Anthropic exports
- Conservative synthetic contradiction generation
- Difficulty splitting for curriculum learning
- Benchmarking tools for external AI contradiction detectors
"""

import csv
import json
import random
import time
import hashlib
from dataclasses import dataclass, asdict
from pathlib import Path
from typing import List, Dict, Any, Optional, Callable


# Optional project imports.
# Kept guarded so this file can still run standalone in test environments.
try:
    from ..core.types import Contradiction, LearningEvent, ContradictionType
except ImportError:
    Contradiction = None
    LearningEvent = None
    ContradictionType = None


# ---------------------------------------------------------------------
# Data classes
# ---------------------------------------------------------------------

@dataclass
class TrainingExample:
    """A single training example for AI contradiction detection."""
    id: str
    premise: str
    query: str
    contradiction_type: str
    explanation: str
    human_patch: str
    repair_action: str
    domain: str
    difficulty_level: int


@dataclass
class AIBenchmarkResult:
    """Results from testing an AI system on contradiction detection."""
    system_name: str
    version: str
    test_date: str
    accuracy: float
    precision_by_type: Dict[str, float]
    recall_by_type: Dict[str, float]
    processing_time: float
    examples_tested: int
    failed_examples: List[Dict[str, str]]


# ---------------------------------------------------------------------
# Utility helpers
# ---------------------------------------------------------------------

def stable_hash(text: str, length: int = 16) -> str:
    """Stable hash for reproducible IDs."""
    return hashlib.sha256(text.encode("utf-8")).hexdigest()[:length]


def normalize_contradiction_type(label: str) -> str:
    """
    Normalize contradiction labels so benchmark comparison is fair.

    Examples:
        "Type II: Property Mismatch" -> "property_mismatch"
        "No contradiction" -> "none"
    """
    if label is None:
        return "unknown"

    raw = str(label).strip()
    key = raw.lower().strip()

    aliases = {
        "type i": "direct_negation",
        "type i: direct negation": "direct_negation",
        "type 1": "direct_negation",
        "type 1: direct negation": "direct_negation",
        "direct negation": "direct_negation",

        "type ii": "property_mismatch",
        "type ii: property mismatch": "property_mismatch",
        "type 2": "property_mismatch",
        "type 2: property mismatch": "property_mismatch",
        "property mismatch": "property_mismatch",

        "type iii": "definitional_violation",
        "type iii: definitional violation": "definitional_violation",
        "type 3": "definitional_violation",
        "type 3: definitional violation": "definitional_violation",
        "definitional violation": "definitional_violation",

        "type iv": "universal_counterexample",
        "type iv: counterexample to universal": "universal_counterexample",
        "type iv: counterexample to universal claim": "universal_counterexample",
        "type 4": "universal_counterexample",
        "type 4: counterexample to universal": "universal_counterexample",
        "counterexample to universal": "universal_counterexample",
        "universal counterexample": "universal_counterexample",

        "type v": "temporal_conflict",
        "type v: temporal conflict": "temporal_conflict",
        "temporal conflict": "temporal_conflict",

        "none": "none",
        "no contradiction": "none",
        "not a contradiction": "none",
        "no logical contradiction": "none",

        "unknown": "unknown",
    }

    return aliases.get(key, key.replace(":", "").replace("-", "_").replace(" ", "_"))


# ---------------------------------------------------------------------
# Training data generator
# ---------------------------------------------------------------------

class ContradictionTrainingDataGenerator:
    """
    Generates training data for AI systems from recursive MBT learning logs.

    Supports:
    - OpenAI chat/completion JSONL
    - Anthropic-style JSON
    - Generic CSV/JSON/TSV
    - Conservative synthetic examples
    - Difficulty-based curriculum splits
    """

    def __init__(self):
        self.training_examples: List[TrainingExample] = []
        self.ai_benchmark_results: List[AIBenchmarkResult] = []

    def process_learning_log(self, learning_log_path: str) -> List[TrainingExample]:
        """
        Convert recursive learning log into training examples.
        """
        with open(learning_log_path, "r", encoding="utf-8") as f:
            log_data = json.load(f)

        examples: List[TrainingExample] = []
        learning_events = log_data.get("learning_events", [])

        contradictions: Dict[str, Dict[str, Any]] = {}
        patches_by_claim_id: Dict[str, Dict[str, Any]] = {}
        patches_by_contradiction_id: Dict[str, Dict[str, Any]] = {}

        for event in learning_events:
            event_type = event.get("event_type")
            data = event.get("data", {})

            if event_type == "contradiction_detection":
                claim_id = data.get("claim_id")
                if claim_id:
                    contradictions[claim_id] = data

            elif event_type == "patch_application":
                claim_id = data.get("claim_id")
                contradiction_id = data.get("contradiction_id")

                if claim_id:
                    patches_by_claim_id[claim_id] = data
                if contradiction_id:
                    patches_by_contradiction_id[contradiction_id] = data

        for claim_id, contradiction_data in contradictions.items():
            stable_contradiction_id = f"contradiction_{stable_hash(claim_id)}"

            patch_data = (
                patches_by_claim_id.get(claim_id)
                or patches_by_contradiction_id.get(stable_contradiction_id)
                or patches_by_contradiction_id.get(f"contradiction_{claim_id}")
                or {}
            )

            raw_type = contradiction_data.get("type", "unknown")
            norm_type = normalize_contradiction_type(raw_type)

            example = TrainingExample(
                id=str(claim_id),
                premise=contradiction_data.get("premise", ""),
                query=contradiction_data.get("query", ""),
                contradiction_type=norm_type,
                explanation=contradiction_data.get("note", ""),
                human_patch=patch_data.get("human_insight", ""),
                repair_action=patch_data.get("repair_action", ""),
                domain=contradiction_data.get("domain", "unknown"),
                difficulty_level=self._assess_difficulty(contradiction_data),
            )

            examples.append(example)

        self.training_examples.extend(examples)
        return examples

    def validate_examples(self) -> Dict[str, Any]:
        """
        Validate examples before export.
        """
        valid_types = {
            "direct_negation",
            "property_mismatch",
            "definitional_violation",
            "universal_counterexample",
            "temporal_conflict",
            "none",
            "unknown",
        }

        report = {
            "total_examples": len(self.training_examples),
            "missing_fields": [],
            "invalid_types": [],
            "duplicate_ids": [],
            "is_valid": True,
        }

        seen_ids = set()

        for example in self.training_examples:
            if example.id in seen_ids:
                report["duplicate_ids"].append(example.id)
            seen_ids.add(example.id)

            missing = []
            for field in ["id", "premise", "query", "contradiction_type"]:
                value = getattr(example, field, "")
                if not str(value).strip():
                    missing.append(field)

            if missing:
                report["missing_fields"].append({
                    "id": example.id,
                    "missing": missing,
                })

            if example.contradiction_type not in valid_types:
                report["invalid_types"].append({
                    "id": example.id,
                    "type": example.contradiction_type,
                })

        report["is_valid"] = (
            not report["missing_fields"]
            and not report["invalid_types"]
            and not report["duplicate_ids"]
        )

        return report

    def export_for_openai_finetuning(
        self,
        output_path: str,
        format_type: str = "chat",
    ) -> None:
        """
        Export training data in OpenAI fine-tuning JSONL format.
        """
        if format_type not in {"chat", "completion"}:
            raise ValueError("format_type must be 'chat' or 'completion'")

        Path(output_path).parent.mkdir(parents=True, exist_ok=True)

        with open(output_path, "w", encoding="utf-8") as f:
            for example in self.training_examples:
                if format_type == "chat":
                    entry = {
                        "messages": [
                            {
                                "role": "system",
                                "content": (
                                    "You are a contradiction detection system. "
                                    "Classify the relationship between the premise and query as one of: "
                                    "direct_negation, property_mismatch, definitional_violation, "
                                    "universal_counterexample, temporal_conflict, none, or unknown. "
                                    "Give a concise explanation and repair action."
                                ),
                            },
                            {
                                "role": "user",
                                "content": (
                                    f"Premise: {example.premise}\n"
                                    f"Query: {example.query}\n\n"
                                    "What type of contradiction is this?"
                                ),
                            },
                            {
                                "role": "assistant",
                                "content": (
                                    f"contradiction_type: {example.contradiction_type}\n"
                                    f"explanation: {example.explanation}\n"
                                    f"repair_action: {example.repair_action}"
                                ),
                            },
                        ]
                    }
                else:
                    entry = {
                        "prompt": (
                            f"Premise: {example.premise}\n"
                            f"Query: {example.query}\n"
                            "Contradiction type:"
                        ),
                        "completion": (
                            f" {example.contradiction_type}. "
                            f"{example.explanation}"
                        ),
                    }

                f.write(json.dumps(entry, ensure_ascii=False) + "\n")

    def export_for_anthropic_training(self, output_path: str) -> None:
        """
        Export training data in a simple Anthropic-style JSON format.
        """
        Path(output_path).parent.mkdir(parents=True, exist_ok=True)

        training_data = []

        for example in self.training_examples:
            entry = {
                "human": (
                    "Analyze this pair of statements for contradiction.\n\n"
                    f"Premise: {example.premise}\n"
                    f"Query: {example.query}\n"
                    f"Domain: {example.domain}\n\n"
                    "Return the contradiction type, explanation, and repair strategy."
                ),
                "assistant": (
                    f"contradiction_type: {example.contradiction_type}\n"
                    f"explanation: {example.explanation}\n"
                    f"human_patch: {example.human_patch}\n"
                    f"repair_action: {example.repair_action}"
                ),
            }

            training_data.append(entry)

        with open(output_path, "w", encoding="utf-8") as f:
            json.dump(training_data, f, indent=2, ensure_ascii=False)

    def export_for_generic_training(self, output_path: str, format: str = "csv") -> None:
        """
        Export in generic CSV, JSON, or TSV.
        """
        if format not in {"csv", "json", "tsv"}:
            raise ValueError("format must be 'csv', 'json', or 'tsv'")

        Path(output_path).parent.mkdir(parents=True, exist_ok=True)

        fields = [
            "id",
            "premise",
            "query",
            "contradiction_type",
            "explanation",
            "human_patch",
            "repair_action",
            "domain",
            "difficulty_level",
        ]

        if format == "csv":
            with open(output_path, "w", newline="", encoding="utf-8") as f:
                writer = csv.writer(f)
                writer.writerow(fields)

                for example in self.training_examples:
                    writer.writerow([getattr(example, field) for field in fields])

        elif format == "json":
            with open(output_path, "w", encoding="utf-8") as f:
                json.dump(
                    [asdict(example) for example in self.training_examples],
                    f,
                    indent=2,
                    ensure_ascii=False,
                )

        elif format == "tsv":
            with open(output_path, "w", encoding="utf-8") as f:
                f.write("\t".join(fields) + "\n")

                for example in self.training_examples:
                    row = [str(getattr(example, field)).replace("\t", " ") for field in fields]
                    f.write("\t".join(row) + "\n")

    def generate_synthetic_examples(
        self,
        base_examples: Optional[List[TrainingExample]] = None,
        count: int = 100,
        seed: Optional[int] = None,
    ) -> List[TrainingExample]:
        """
        Generate conservative synthetic examples from vetted templates.

        This avoids random cross-domain slot filling.
        """
        if seed is not None:
            random.seed(seed)

        template_bank = [
            {
                "premise": "Water boils at 100C at sea level.",
                "query": "Water boils at 95C at sea level.",
                "contradiction_type": "property_mismatch",
                "explanation": "The two statements assign different boiling points to water under the same condition.",
                "repair_action": "Specify different pressure conditions or correct one of the temperature values.",
                "domain": "physics",
                "difficulty_level": 2,
            },
            {
                "premise": "A triangle has exactly three sides.",
                "query": "This triangle has four sides.",
                "contradiction_type": "definitional_violation",
                "explanation": "Having four sides violates the definition of a triangle.",
                "repair_action": "Rename the object as a quadrilateral or correct the side count.",
                "domain": "geometry",
                "difficulty_level": 3,
            },
            {
                "premise": "All birds can fly.",
                "query": "Penguins are birds that cannot fly.",
                "contradiction_type": "universal_counterexample",
                "explanation": "Penguins are a counterexample to the universal claim that all birds can fly.",
                "repair_action": "Revise the universal claim to allow exceptions.",
                "domain": "biology",
                "difficulty_level": 4,
            },
            {
                "premise": "The meeting starts at 10:00.",
                "query": "The same meeting starts at 11:00.",
                "contradiction_type": "temporal_conflict",
                "explanation": "The same event is assigned two different start times.",
                "repair_action": "Clarify whether one time is updated, mistaken, or refers to a different timezone.",
                "domain": "scheduling",
                "difficulty_level": 3,
            },
            {
                "premise": "The door is open.",
                "query": "The door is not open.",
                "contradiction_type": "direct_negation",
                "explanation": "The query directly negates the premise.",
                "repair_action": "Clarify time, perspective, or correct one of the statements.",
                "domain": "general",
                "difficulty_level": 1,
            },
            {
                "premise": "Some mammals lay eggs.",
                "query": "Platypuses are mammals that lay eggs.",
                "contradiction_type": "none",
                "explanation": "The query is consistent with the premise.",
                "repair_action": "No repair needed.",
                "domain": "biology",
                "difficulty_level": 2,
            },
        ]

        surface_forms = [
            ("Premise: {premise}", "Query: {query}"),
            ("Assume this is true: {premise}", "Can this also be true: {query}"),
            ("Statement A: {premise}", "Statement B: {query}"),
            ("Known claim: {premise}", "New claim: {query}"),
        ]

        synthetic_examples: List[TrainingExample] = []

        for i in range(count):
            item = random.choice(template_bank)
            premise_form, query_form = random.choice(surface_forms)

            premise = premise_form.format(premise=item["premise"])
            query = query_form.format(query=item["query"])

            synthetic_examples.append(
                TrainingExample(
                    id=f"synthetic_{stable_hash(premise + query, 12)}_{i:04d}",
                    premise=premise,
                    query=query,
                    contradiction_type=item["contradiction_type"],
                    explanation=item["explanation"],
                    human_patch="Synthetic vetted template.",
                    repair_action=item["repair_action"],
                    domain=item["domain"],
                    difficulty_level=item["difficulty_level"],
                )
            )

        self.training_examples.extend(synthetic_examples)
        return synthetic_examples

    def create_difficulty_split(
        self,
        test_split: float = 0.2,
        seed: Optional[int] = 42,
    ) -> Dict[str, List[TrainingExample]]:
        """
        Split examples by difficulty for curriculum learning.
        """
        if not 0.0 <= test_split <= 1.0:
            raise ValueError("test_split must be between 0 and 1")

        rng = random.Random(seed)

        easy = [ex for ex in self.training_examples if ex.difficulty_level <= 2]
        medium = [ex for ex in self.training_examples if ex.difficulty_level == 3]
        hard = [ex for ex in self.training_examples if ex.difficulty_level >= 4]

        rng.shuffle(easy)
        rng.shuffle(medium)
        rng.shuffle(hard)

        easy_test_size = int(len(easy) * test_split)
        medium_test_size = int(len(medium) * test_split)
        hard_test_size = int(len(hard) * test_split)

        return {
            "easy": easy[easy_test_size:],
            "medium": medium[medium_test_size:],
            "hard": hard[hard_test_size:],
            "test": easy[:easy_test_size]
            + medium[:medium_test_size]
            + hard[:hard_test_size],
        }

    def _assess_difficulty(self, contradiction_data: Dict[str, Any]) -> int:
        """
        Assess difficulty level of a contradiction on a 1-5 scale.
        """
        contradiction_type = normalize_contradiction_type(
            contradiction_data.get("type", "unknown")
        )

        premise = contradiction_data.get("premise", "")
        query = contradiction_data.get("query", "")
        note = contradiction_data.get("note", "")

        base = {
            "direct_negation": 1,
            "property_mismatch": 2,
            "definitional_violation": 3,
            "temporal_conflict": 3,
            "universal_counterexample": 4,
            "none": 2,
            "unknown": 3,
        }.get(contradiction_type, 3)

        word_count = len(str(premise).split()) + len(str(query).split())

        if word_count > 40:
            base += 1
        if len(str(note).split()) > 40:
            base += 1

        return max(1, min(5, base))


# ---------------------------------------------------------------------
# Benchmarking
# ---------------------------------------------------------------------

class AISystemBenchmarker:
    """
    Benchmarking framework for contradiction detection systems.

    prediction_function must take:
        (premise: str, query: str) -> str
    """

    def __init__(self):
        self.benchmark_results: List[AIBenchmarkResult] = []
        self.test_suite: List[TrainingExample] = []

    def load_test_suite(self, test_examples: List[TrainingExample]) -> None:
        self.test_suite = test_examples

    def benchmark_ai_system(
        self,
        system_name: str,
        system_version: str,
        prediction_function: Callable[[str, str], str],
        test_examples: Optional[List[TrainingExample]] = None,
    ) -> AIBenchmarkResult:
        """
        Benchmark contradiction detection performance.
        """
        if test_examples is None:
            test_examples = self.test_suite

        if not test_examples:
            raise ValueError("No test examples available")

        correct_predictions = 0
        predictions_by_type: Dict[str, List[str]] = {}
        actual_by_type: Dict[str, List[str]] = {}
        processing_times: List[float] = []
        failed_examples: List[Dict[str, str]] = []

        for example in test_examples:
            start_time = time.time()

            try:
                raw_predicted_type = prediction_function(example.premise, example.query)
                predicted_type = normalize_contradiction_type(raw_predicted_type)
                actual_type = normalize_contradiction_type(example.contradiction_type)

                processing_times.append(time.time() - start_time)

                predictions_by_type.setdefault(predicted_type, []).append(actual_type)
                actual_by_type.setdefault(actual_type, []).append(predicted_type)

                if predicted_type == actual_type:
                    correct_predictions += 1

            except Exception as e:
                processing_times.append(time.time() - start_time)
                failed_examples.append({
                    "id": example.id,
                    "error": str(e),
                })

        total = len(test_examples)
        accuracy = correct_predictions / total if total else 0.0

        precision_by_type: Dict[str, float] = {}
        recall_by_type: Dict[str, float] = {}

        for pred_type, actual_types in predictions_by_type.items():
            true_positives = sum(1 for actual in actual_types if actual == pred_type)
            precision_by_type[pred_type] = (
                true_positives / len(actual_types) if actual_types else 0.0
            )

        for actual_type, pred_types in actual_by_type.items():
            true_positives = sum(1 for pred in pred_types if pred == actual_type)
            recall_by_type[actual_type] = (
                true_positives / len(pred_types) if pred_types else 0.0
            )

        result = AIBenchmarkResult(
            system_name=system_name,
            version=system_version,
            test_date=time.strftime("%Y-%m-%dT%H:%M:%SZ", time.gmtime()),
            accuracy=accuracy,
            precision_by_type=precision_by_type,
            recall_by_type=recall_by_type,
            processing_time=(
                sum(processing_times) / len(processing_times)
                if processing_times else 0.0
            ),
            examples_tested=total,
            failed_examples=failed_examples,
        )

        self.benchmark_results.append(result)
        return result

    def compare_systems(self, results: Optional[List[AIBenchmarkResult]] = None) -> Dict[str, Any]:
        """
        Compare multiple AI systems.
        """
        if results is None:
            results = self.benchmark_results

        comparison: Dict[str, Any] = {
            "systems": [],
            "accuracy_ranking": [],
            "speed_ranking": [],
            "best_by_type": {},
            "summary": {},
        }

        for result in results:
            comparison["systems"].append({
                "name": result.system_name,
                "version": result.version,
                "accuracy": result.accuracy,
                "avg_processing_time": result.processing_time,
                "examples_tested": result.examples_tested,
                "failed_examples": len(result.failed_examples),
            })

        comparison["accuracy_ranking"] = [
            {
                "system": r.system_name,
                "version": r.version,
                "accuracy": r.accuracy,
            }
            for r in sorted(results, key=lambda x: x.accuracy, reverse=True)
        ]

        comparison["speed_ranking"] = [
            {
                "system": r.system_name,
                "version": r.version,
                "avg_processing_time": r.processing_time,
            }
            for r in sorted(results, key=lambda x: x.processing_time)
        ]

        all_types = set()
        for result in results:
            all_types.update(result.precision_by_type.keys())
            all_types.update(result.recall_by_type.keys())

        for ctype in sorted(all_types):
            best_precision = max(
                results,
                key=lambda x: x.precision_by_type.get(ctype, 0.0),
            )
            best_recall = max(
                results,
                key=lambda x: x.recall_by_type.get(ctype, 0.0),
            )

            comparison["best_by_type"][ctype] = {
                "best_precision": {
                    "system": best_precision.system_name,
                    "score": best_precision.precision_by_type.get(ctype, 0.0),
                },
                "best_recall": {
                    "system": best_recall.system_name,
                    "score": best_recall.recall_by_type.get(ctype, 0.0),
                },
            }

        if results:
            comparison["summary"] = {
                "systems_tested": len(results),
                "best_accuracy": max(r.accuracy for r in results),
                "fastest_avg_time": min(r.processing_time for r in results),
            }

        return comparison

    def export_benchmark_results(self, output_path: str) -> None:
        """
        Export benchmark results to JSON.
        """
        Path(output_path).parent.mkdir(parents=True, exist_ok=True)

        results_data = {
            "benchmark_results": [
                asdict(result) for result in self.benchmark_results
            ],
            "export_timestamp": time.strftime("%Y-%m-%dT%H:%M:%SZ", time.gmtime()),
            "total_systems_tested": len(self.benchmark_results),
        }

        with open(output_path, "w", encoding="utf-8") as f:
            json.dump(results_data, f, indent=2, ensure_ascii=False)


# ---------------------------------------------------------------------
# Convenience functions
# ---------------------------------------------------------------------

def create_training_data_from_logs(
    learning_log_paths: List[str],
    output_dir: str = "training_data",
    synthetic_count: int = 200,
    seed: int = 42,
) -> Dict[str, Any]:
    """
    Create training exports from one or more MBT learning logs.
    """
    Path(output_dir).mkdir(parents=True, exist_ok=True)

    generator = ContradictionTrainingDataGenerator()

    for log_path in learning_log_paths:
        print(f"Processing {log_path}...")
        generator.process_learning_log(log_path)

    print(f"Generated {len(generator.training_examples)} extracted training examples")

    synthetic = generator.generate_synthetic_examples(
        count=synthetic_count,
        seed=seed,
    )

    print(f"Generated {len(synthetic)} synthetic examples")

    validation_report = generator.validate_examples()

    with open(f"{output_dir}/validation_report.json", "w", encoding="utf-8") as f:
        json.dump(validation_report, f, indent=2, ensure_ascii=False)

    generator.export_for_openai_finetuning(
        f"{output_dir}/openai_chat_format.jsonl",
        "chat",
    )
    generator.export_for_openai_finetuning(
        f"{output_dir}/openai_completion_format.jsonl",
        "completion",
    )
    generator.export_for_anthropic_training(
        f"{output_dir}/anthropic_format.json",
    )
    generator.export_for_generic_training(
        f"{output_dir}/training_data.csv",
        "csv",
    )
    generator.export_for_generic_training(
        f"{output_dir}/training_data.json",
        "json",
    )
    generator.export_for_generic_training(
        f"{output_dir}/training_data.tsv",
        "tsv",
    )

    splits = generator.create_difficulty_split(seed=seed)

    for split_name, examples in splits.items():
        with open(f"{output_dir}/{split_name}_examples.json", "w", encoding="utf-8") as f:
            json.dump([asdict(ex) for ex in examples], f, indent=2, ensure_ascii=False)

    print(f"Training data exported to {output_dir}/")
    print(f"Validation passed: {validation_report['is_valid']}")

    return {
        "total_examples": len(generator.training_examples),
        "synthetic_examples": len(synthetic),
        "validation": validation_report,
        "output_dir": output_dir,
    }


def benchmark_against_mbt(
    mbt_results_path: str,
    ai_prediction_function: Callable[[str, str], str],
    system_name: str,
    system_version: str = "1.0",
) -> AIBenchmarkResult:
    """
    Benchmark an AI system against MBT result labels.
    """
    with open(mbt_results_path, "r", encoding="utf-8") as f:
        mbt_data = json.load(f)

    test_examples: List[TrainingExample] = []

    for result in mbt_data.get("results", []):
        if result.get("status") == "cycle_complete":
            contradiction = result.get("contradiction", {})

            example = TrainingExample(
                id=str(result.get("claim_id", stable_hash(json.dumps(result, sort_keys=True)))),
                premise=contradiction.get("premise", ""),
                query=contradiction.get("query", ""),
                contradiction_type=normalize_contradiction_type(
                    contradiction.get("type", "unknown")
                ),
                explanation=contradiction.get("note", ""),
                human_patch="",
                repair_action="",
                domain=contradiction.get("domain", "unknown"),
                difficulty_level=3,
            )

            test_examples.append(example)

    benchmarker = AISystemBenchmarker()
    benchmarker.load_test_suite(test_examples)

    result = benchmarker.benchmark_ai_system(
        system_name=system_name,
        system_version=system_version,
        prediction_function=ai_prediction_function,
        test_examples=test_examples,
    )

    print(f"\nBenchmark Results for {system_name} v{system_version}")
    print(f"Accuracy: {result.accuracy:.2%}")
    print(f"Processing time: {result.processing_time:.4f}s per example")
    print(f"Examples tested: {result.examples_tested}")
    print(f"Failed examples: {len(result.failed_examples)}")

    return result


# ---------------------------------------------------------------------
# Optional demo
# ---------------------------------------------------------------------

if __name__ == "__main__":
    generator = ContradictionTrainingDataGenerator()
    generator.generate_synthetic_examples(count=20, seed=42)

    report = generator.validate_examples()
    print(json.dumps(report, indent=2))

    splits = generator.create_difficulty_split()
    print({
        key: len(value)
        for key, value in splits.items()
    })
````

````python
# Install once if needed:
# pip install datasets

from datasets import load_dataset
from dataclasses import asdict
import json
from collections import defaultdict


def f1_from_precision_recall(p, r):
    if p + r == 0:
        return 0.0
    return 2 * p * r / (p + r)


def add_f1_report(result):
    all_types = set(result.precision_by_type) | set(result.recall_by_type)

    f1_by_type = {}
    for ctype in sorted(all_types):
        p = result.precision_by_type.get(ctype, 0.0)
        r = result.recall_by_type.get(ctype, 0.0)
        f1_by_type[ctype] = f1_from_precision_recall(p, r)

    macro_f1 = sum(f1_by_type.values()) / len(f1_by_type) if f1_by_type else 0.0

    return {
        "system_name": result.system_name,
        "version": result.version,
        "accuracy": result.accuracy,
        "macro_f1": macro_f1,
        "precision_by_type": result.precision_by_type,
        "recall_by_type": result.recall_by_type,
        "f1_by_type": f1_by_type,
        "examples_tested": result.examples_tested,
        "failed_examples": len(result.failed_examples),
    }


def load_snli_examples(limit=300, split="validation", seed=42):
    """
    Pull SNLI from Hugging Face and convert to TrainingExample objects.

    SNLI labels:
      0 = entailment
      1 = neutral
      2 = contradiction
      -1 = invalid / no gold label
    """
    ds = load_dataset("stanfordnlp/snli", split=split)
    ds = ds.shuffle(seed=seed)

    examples = []
    label_map = {
        0: "none",          # entailment is non-contradiction for this benchmark
        1: "none",          # neutral is also non-contradiction here
        2: "direct_negation" # SNLI only says contradiction, not MBT subtype
    }

    for row in ds:
        label = row["label"]
        if label == -1:
            continue

        ctype = label_map[label]

        ex = TrainingExample(
            id=f"snli_{split}_{len(examples):06d}",
            premise=row["premise"],
            query=row["hypothesis"],
            contradiction_type=ctype,
            explanation=f"SNLI gold label mapped from label={label}.",
            human_patch="",
            repair_action="",
            domain="snli_natural_language_inference",
            difficulty_level=3,
        )

        examples.append(ex)

        if len(examples) >= limit:
            break

    return examples


snli_examples = load_snli_examples(limit=300, split="validation", seed=42)

print("Loaded SNLI examples:", len(snli_examples))
print("Label counts:")
counts = defaultdict(int)
for ex in snli_examples:
    counts[ex.contradiction_type] += 1
print(dict(counts))


snli_benchmarker = AISystemBenchmarker()
snli_benchmarker.load_test_suite(snli_examples)


def snli_keyword_baseline(premise, query):
    """
    Intentionally weak baseline for external natural-language data.
    This should not perform like the synthetic benchmark.
    """
    text = f"{premise} {query}".lower()

    negation_words = [" not ", " never ", " no ", " nobody ", " nothing ", "n't"]
    if any(w in text for w in negation_words):
        return "direct_negation"

    return "none"


snli_result = snli_benchmarker.benchmark_ai_system(
    system_name="SNLIKeywordBaseline",
    system_version="negation_heuristic_v1",
    prediction_function=snli_keyword_baseline,
)

report = add_f1_report(snli_result)
print(json.dumps(report, indent=2))
````

````python
from collections import Counter, defaultdict
import json

def confusion_matrix_report(examples, predictor):
    matrix = defaultdict(Counter)

    for ex in examples:
        actual = ex.contradiction_type
        pred = predictor(ex.premise, ex.query)
        matrix[actual][pred] += 1

    return {
        actual: dict(preds)
        for actual, preds in matrix.items()
    }


def always_none(premise, query):
    return "none"


def run_and_report(name, version, predictor):
    result = snli_benchmarker.benchmark_ai_system(
        system_name=name,
        system_version=version,
        prediction_function=predictor,
    )

    report = add_f1_report(result)
    report["confusion_matrix"] = confusion_matrix_report(snli_examples, predictor)

    print("\n" + name, version)
    print(json.dumps(report, indent=2))

    return result, report


r_none, rep_none = run_and_report(
    "SNLIBaseline",
    "always_none",
    always_none,
)

r_neg, rep_neg = run_and_report(
    "SNLIBaseline",
    "negation_only",
    snli_keyword_baseline,
)

r_sem, rep_sem = run_and_report(
    "SNLIBaseline",
    "semantic_pairs",
    snli_keyword_baseline_v2,
)


full_report = {
    "dataset": "SNLI validation sample",
    "n_examples": len(snli_examples),
    "label_counts": dict(Counter(ex.contradiction_type for ex in snli_examples)),
    "reports": [rep_none, rep_neg, rep_sem],
}

print("\nFULL BENCHMARK REPORT")
print(json.dumps(full_report, indent=2))
````

````python
import re, json, time
from collections import defaultdict

def run_gemini_batches(examples, batch_size=10):
    preds = []

    for i in range(0, len(examples), batch_size):
        batch = examples[i:i+batch_size]

        items = []
        for j, ex in enumerate(batch):
            items.append(
                f"{j}. Premise: {ex.premise}\nHypothesis: {ex.query}\n"
            )

        prompt = """
Classify each pair.

Return JSON only like:
{"0":"C","1":"N"}

C = contradiction
N = no contradiction

Pairs:
""" + "\n".join(items)

        response = gemini_client.models.generate_content(
            model="gemini-2.5-flash-lite",
            contents=prompt,
            config=types.GenerateContentConfig(
                temperature=0,
                max_output_tokens=300,
            ),
        )

        raw = response.text.strip()
        raw = re.sub(r"^```json\s*|\s*```$", "", raw).strip()

        try:
            batch_preds = json.loads(raw)
        except:
            print("Parse error:", raw)
            batch_preds = {}

        for j, ex in enumerate(batch):
            label = batch_preds.get(str(j), "N")
            pred = "broad_semantic_contradiction" if label == "C" else "none"
            preds.append((ex, pred))

        print(f"Processed {i+len(batch)}/{len(examples)}")

        time.sleep(10)  # avoid rate limit

    return preds


def evaluate(preds):
    tp = fp = fn = tn = 0

    for ex, pred in preds:
        actual = ex.contradiction_type

        if actual == "broad_semantic_contradiction":
            if pred == actual:
                tp += 1
            else:
                fn += 1
        else:
            if pred == "broad_semantic_contradiction":
                fp += 1
            else:
                tn += 1

    accuracy = (tp + tn) / (tp + tn + fp + fn)

    precision = tp / (tp + fp) if (tp + fp) else 0
    recall = tp / (tp + fn) if (tp + fn) else 0
    f1 = 2 * precision * recall / (precision + recall) if (precision + recall) else 0

    return {
        "accuracy": accuracy,
        "precision": precision,
        "recall": recall,
        "f1": f1,
        "tp": tp,
        "fp": fp,
        "fn": fn,
        "tn": tn,
    }


snli_100 = snli_examples[:100]

preds = run_gemini_batches(snli_100)
metrics = evaluate(preds)

print(json.dumps(metrics, indent=2))
````

````python
import re, json, time

gemini_batch_preds = []  # keep this global so you can resume if interrupted

def run_gemini_batches_resumable(examples, batch_size=10, start_batch=0):
    global gemini_batch_preds

    for i in range(start_batch * batch_size, len(examples), batch_size):
        batch = examples[i:i+batch_size]

        items = []
        for j, ex in enumerate(batch):
            items.append(
                f"{j}. Premise: {ex.premise}\nHypothesis: {ex.query}\n"
            )

        prompt = """
Classify each pair.

Return JSON only like:
{"0":"C","1":"N"}

C = contradiction
N = no contradiction

Pairs:
""" + "\n".join(items)

        while True:
            try:
                response = gemini_client.models.generate_content(
                    model="gemini-2.5-flash-lite",
                    contents=prompt,
                    config=types.GenerateContentConfig(
                        temperature=0,
                        max_output_tokens=300,
                    ),
                )

                raw = response.text.strip() if response.text else ""
                raw = re.sub(r"^```json\s*|\s*```$", "", raw).strip()

                batch_preds = json.loads(raw)

                for j, ex in enumerate(batch):
                    label = batch_preds.get(str(j), "N")
                    pred = "broad_semantic_contradiction" if label == "C" else "none"
                    gemini_batch_preds.append((ex, pred))

                print(f"Processed {i+len(batch)}/{len(examples)}")
                time.sleep(12)
                break

            except Exception as e:
                msg = str(e)

                if "429" in msg or "RESOURCE_EXHAUSTED" in msg:
                    print("429 rate limit. Sleeping 30s...")
                    time.sleep(30)
                    continue

                if "503" in msg or "UNAVAILABLE" in msg:
                    print("503 high demand. Sleeping 45s...")
                    time.sleep(45)
                    continue

                print("Non-retry error:", e)
                for ex in batch:
                    gemini_batch_preds.append((ex, "unknown"))
                break

    return gemini_batch_preds
````

````python
def mbt_subclass_predictor_v3(premise, query):
    text = f"{premise} {query}".lower()

    # 1. Direct negation (highest priority)
    negation_words = [" not ", " never ", " no ", " nobody ", " nothing ", "n't"]
    if any(w in text for w in negation_words):
        return "direct_negation"

    # 2. Unsupported mental state → NONE
    mental_words = ["fear", "hate", "love", "thinking", "feels", "believes"]
    if any(w in text for w in mental_words):
        return "none"

    # 3. Property mismatch = single attribute clash
    attribute_pairs = [
        ("sitting", "standing"),
        ("standing", "sitting"),
        ("boy", "girl"),
        ("man", "woman"),
        ("black", "white"),
        ("gray", "black"),
        ("grey", "black"),
        ("one", "two"),
        ("two", "three"),
    ]

    for a, b in attribute_pairs:
        if a in text and b in text:
            return "property_mismatch"

    # 4. Scene replacement → broad semantic contradiction
    # crude heuristic: multiple distinct action/location words
    scene_words = ["inside", "outside", "store", "field", "playground", "home"]
    action_words = ["running", "playing", "working", "sleeping", "shopping"]

    scene_hits = sum(w in text for w in scene_words)
    action_hits = sum(w in text for w in action_words)

    if scene_hits >= 2 or action_hits >= 2:
        return "broad_semantic_contradiction"

    return "broad_semantic_contradiction"
````

````python
# ============================================================
# MBT CONTRADICTION BENCHMARK — ONE-CELL CLEAN RESTART
# ============================================================
# What this does:
# 1. Installs dependencies
# 2. Loads SNLI validation data from Hugging Face
# 3. Converts SNLI labels into binary contradiction labels
# 4. Runs rule baselines
# 5. Runs Gemini in safe batches with retry/backoff
# 6. Computes accuracy, precision, recall, F1, macro-F1, confusion matrix
# 7. Saves all reports to JSON files
#
# No editing required.
# ============================================================

import sys
import subprocess
import importlib.util

def ensure_package(package_name, import_name=None):
    import_name = import_name or package_name
    if importlib.util.find_spec(import_name) is None:
        subprocess.check_call([sys.executable, "-m", "pip", "install", "-q", package_name])

ensure_package("datasets")
ensure_package("google-genai", "google.genai")

import os
import re
import json
import time
import random
from dataclasses import dataclass, asdict
from typing import List, Dict, Any, Callable, Optional
from collections import Counter, defaultdict
from getpass import getpass

from datasets import load_dataset
from google import genai
from google.genai import types


# ============================================================
# CONFIG
# ============================================================

SNLI_SPLIT = "validation"
SNLI_LIMIT = 300
SNLI_SEED = 42

GEMINI_MODEL = "gemini-2.5-flash-lite"
GEMINI_BATCH_SIZE = 10
GEMINI_LIMIT = 100          # change later if wanted, but no need now
GEMINI_SLEEP_SECONDS = 12
GEMINI_MAX_RETRIES = 8

OUTPUT_PREFIX = "mbt_snli_clean"


# ============================================================
# DATA TYPES
# ============================================================

@dataclass
class TrainingExample:
    id: str
    premise: str
    query: str
    contradiction_type: str
    explanation: str
    human_patch: str
    repair_action: str
    domain: str
    difficulty_level: int


@dataclass
class BenchmarkResult:
    system_name: str
    version: str
    accuracy: float
    precision_by_type: Dict[str, float]
    recall_by_type: Dict[str, float]
    f1_by_type: Dict[str, float]
    macro_f1: float
    confusion_matrix: Dict[str, Dict[str, int]]
    examples_tested: int
    failed_examples: int
    errors: List[Dict[str, Any]]


# ============================================================
# METRICS
# ============================================================

def f1_score(p, r):
    if p + r == 0:
        return 0.0
    return 2 * p * r / (p + r)


def evaluate_predictions(
    examples: List[TrainingExample],
    predictions: List[str],
    system_name: str,
    version: str,
) -> BenchmarkResult:
    if len(examples) != len(predictions):
        raise ValueError(f"examples/predictions length mismatch: {len(examples)} vs {len(predictions)}")

    labels = sorted(set([ex.contradiction_type for ex in examples] + predictions))

    matrix = defaultdict(Counter)
    errors = []

    for ex, pred in zip(examples, predictions):
        actual = ex.contradiction_type
        matrix[actual][pred] += 1

        if pred != actual:
            errors.append({
                "id": ex.id,
                "actual": actual,
                "predicted": pred,
                "premise": ex.premise,
                "query": ex.query,
            })

    correct = sum(1 for ex, pred in zip(examples, predictions) if ex.contradiction_type == pred)
    accuracy = correct / len(examples) if examples else 0.0

    precision_by_type = {}
    recall_by_type = {}
    f1_by_type = {}

    for label in labels:
        tp = matrix[label][label]
        predicted_as_label = sum(matrix[actual][label] for actual in matrix)
        actual_label = sum(matrix[label].values())

        precision = tp / predicted_as_label if predicted_as_label else 0.0
        recall = tp / actual_label if actual_label else 0.0
        f1 = f1_score(precision, recall)

        precision_by_type[label] = precision
        recall_by_type[label] = recall
        f1_by_type[label] = f1

    macro_f1 = sum(f1_by_type.values()) / len(f1_by_type) if f1_by_type else 0.0

    return BenchmarkResult(
        system_name=system_name,
        version=version,
        accuracy=accuracy,
        precision_by_type=precision_by_type,
        recall_by_type=recall_by_type,
        f1_by_type=f1_by_type,
        macro_f1=macro_f1,
        confusion_matrix={actual: dict(preds) for actual, preds in matrix.items()},
        examples_tested=len(examples),
        failed_examples=sum(1 for p in predictions if p == "unknown"),
        errors=errors,
    )


def print_result(result: BenchmarkResult):
    print("\n" + "=" * 80)
    print(f"{result.system_name} — {result.version}")
    print("=" * 80)
    print(f"Accuracy:        {result.accuracy:.4f}")
    print(f"Macro-F1:        {result.macro_f1:.4f}")
    print(f"Examples tested: {result.examples_tested}")
    print(f"Failed/unknown:  {result.failed_examples}")
    print("\nPrecision by type:")
    print(json.dumps(result.precision_by_type, indent=2))
    print("\nRecall by type:")
    print(json.dumps(result.recall_by_type, indent=2))
    print("\nF1 by type:")
    print(json.dumps(result.f1_by_type, indent=2))
    print("\nConfusion matrix:")
    print(json.dumps(result.confusion_matrix, indent=2))
    print("\nErrors shown:", min(10, len(result.errors)), "of", len(result.errors))

    for err in result.errors[:10]:
        print("\n--- ERROR ---")
        print("ID:", err["id"])
        print("Actual:", err["actual"])
        print("Predicted:", err["predicted"])
        print("Premise:", err["premise"])
        print("Query:", err["query"])


# ============================================================
# LOAD SNLI
# ============================================================

def load_snli_binary(limit=300, split="validation", seed=42) -> List[TrainingExample]:
    """
    SNLI labels:
      0 = entailment
      1 = neutral
      2 = contradiction
      -1 = invalid

    Mapping used here:
      contradiction -> broad_semantic_contradiction
      entailment/neutral -> none
    """
    ds = load_dataset("stanfordnlp/snli", split=split)
    ds = ds.shuffle(seed=seed)

    examples = []

    for row in ds:
        label = row["label"]
        if label == -1:
            continue

        if label == 2:
            contradiction_type = "broad_semantic_contradiction"
        else:
            contradiction_type = "none"

        examples.append(
            TrainingExample(
                id=f"snli_{split}_{len(examples):06d}",
                premise=row["premise"],
                query=row["hypothesis"],
                contradiction_type=contradiction_type,
                explanation=f"SNLI label={label} mapped to {contradiction_type}",
                human_patch="",
                repair_action="",
                domain="snli_binary",
                difficulty_level=3,
            )
        )

        if len(examples) >= limit:
            break

    return examples


snli_examples = load_snli_binary(
    limit=SNLI_LIMIT,
    split=SNLI_SPLIT,
    seed=SNLI_SEED,
)

print("\nLoaded SNLI examples:", len(snli_examples))
print("Label counts:", dict(Counter(ex.contradiction_type for ex in snli_examples)))


# ============================================================
# RULE BASELINES
# ============================================================

def always_none_predictor(ex: TrainingExample) -> str:
    return "none"


def negation_predictor(ex: TrainingExample) -> str:
    text = f" {ex.premise} {ex.query} ".lower()
    negation_words = [" not ", " never ", " no ", " nobody ", " nothing ", "n't"]
    if any(w in text for w in negation_words):
        return "broad_semantic_contradiction"
    return "none"


def semantic_pairs_predictor(ex: TrainingExample) -> str:
    text = f" {ex.premise} {ex.query} ".lower()

    contradiction_pairs = [
        ("sitting", "standing"),
        ("standing", "sitting"),
        ("running", "stationary"),
        ("racing", "stationary"),
        ("inside", "outside"),
        ("outside", "inside"),
        ("boy", "girl"),
        ("girl", "boy"),
        ("man", "woman"),
        ("woman", "man"),
        ("sleeping", "shopping"),
        ("shopping", "sleeping"),
        ("sleeping", "playing"),
        ("playing", "sleeping"),
        ("asleep", "cleaning"),
        ("cleaning", "asleep"),
        ("gray", "black"),
        ("grey", "black"),
        ("black", "gray"),
        ("black", "grey"),
    ]

    negation_words = [" not ", " never ", " no ", " nobody ", " nothing ", "n't"]
    if any(w in text for w in negation_words):
        return "broad_semantic_contradiction"

    for a, b in contradiction_pairs:
        if a in text and b in text:
            return "broad_semantic_contradiction"

    return "none"


def run_rule_baseline(name, version, predictor, examples):
    predictions = [predictor(ex) for ex in examples]
    result = evaluate_predictions(examples, predictions, name, version)
    print_result(result)
    return result


baseline_results = []

baseline_results.append(
    run_rule_baseline("RuleBaseline", "always_none", always_none_predictor, snli_examples)
)

baseline_results.append(
    run_rule_baseline("RuleBaseline", "negation_only", negation_predictor, snli_examples)
)

baseline_results.append(
    run_rule_baseline("RuleBaseline", "semantic_pairs", semantic_pairs_predictor, snli_examples)
)


# ============================================================
# GEMINI BATCH EVALUATOR
# ============================================================

def setup_gemini_client():
    if "GEMINI_API_KEY" not in os.environ:
        os.environ["GEMINI_API_KEY"] = getpass("Paste Gemini API key: ")
    return genai.Client(api_key=os.environ["GEMINI_API_KEY"])


gemini_client = setup_gemini_client()


def clean_json_response(text: str) -> str:
    text = text.strip()
    text = re.sub(r"^```json\s*", "", text)
    text = re.sub(r"^```\s*", "", text)
    text = re.sub(r"\s*```$", "", text)
    return text.strip()


def parse_gemini_batch_response(text: str, batch_len: int) -> Dict[str, str]:
    """
    Expected:
      {"0":"C","1":"N"}
    """
    cleaned = clean_json_response(text)

    try:
        data = json.loads(cleaned)
    except Exception:
        print("\nGemini parse failed. Raw text:")
        print(repr(text))
        return {str(i): "unknown" for i in range(batch_len)}

    parsed = {}

    for i in range(batch_len):
        raw = str(data.get(str(i), "")).strip().upper()

        if raw.startswith("C"):
            parsed[str(i)] = "broad_semantic_contradiction"
        elif raw.startswith("N"):
            parsed[str(i)] = "none"
        else:
            parsed[str(i)] = "unknown"

    return parsed


def gemini_batch_predict(
    examples: List[TrainingExample],
    model: str = GEMINI_MODEL,
    batch_size: int = GEMINI_BATCH_SIZE,
    sleep_seconds: int = GEMINI_SLEEP_SECONDS,
    max_retries: int = GEMINI_MAX_RETRIES,
) -> List[str]:

    all_predictions = []

    for start in range(0, len(examples), batch_size):
        batch = examples[start:start + batch_size]

        items = []
        for j, ex in enumerate(batch):
            items.append(
                f"{j}. Premise: {ex.premise}\nHypothesis: {ex.query}\n"
            )

        prompt = """
Classify each pair.

Return JSON only, exactly like:
{"0":"C","1":"N"}

C = contradiction
N = no contradiction

Rules:
- Unsupported extra detail is N.
- A mental state not visible from the premise is usually N.
- Direct conflict in person, action, count, color, location, or scene is C.

Pairs:
""" + "\n".join(items)

        attempt = 0

        while True:
            try:
                response = gemini_client.models.generate_content(
                    model=model,
                    contents=prompt,
                    config=types.GenerateContentConfig(
                        temperature=0,
                        max_output_tokens=500,
                    ),
                )

                raw_text = response.text if response.text else ""
                parsed = parse_gemini_batch_response(raw_text, len(batch))

                for j in range(len(batch)):
                    all_predictions.append(parsed[str(j)])

                print(f"Gemini processed {start + len(batch)}/{len(examples)}")
                time.sleep(sleep_seconds)
                break

            except Exception as e:
                attempt += 1
                msg = str(e)

                if attempt > max_retries:
                    print(f"Gemini failed batch starting at {start}. Marking unknown.")
                    print("Error:", e)
                    all_predictions.extend(["unknown"] * len(batch))
                    break

                if "429" in msg or "RESOURCE_EXHAUSTED" in msg:
                    wait = 30 + attempt * 10
                    print(f"429 rate limit. Sleeping {wait}s. Attempt {attempt}/{max_retries}")
                    time.sleep(wait)
                    continue

                if "503" in msg or "UNAVAILABLE" in msg:
                    wait = 45 + attempt * 15
                    print(f"503 high demand. Sleeping {wait}s. Attempt {attempt}/{max_retries}")
                    time.sleep(wait)
                    continue

                wait = 20 + attempt * 10
                print(f"Gemini error. Sleeping {wait}s. Attempt {attempt}/{max_retries}")
                print("Error:", e)
                time.sleep(wait)

    return all_predictions


print("\nStarting Gemini benchmark.")
print(f"Gemini limit: {GEMINI_LIMIT}")
print(f"Batch size: {GEMINI_BATCH_SIZE}")
print(f"Model: {GEMINI_MODEL}")

gemini_examples = snli_examples[:GEMINI_LIMIT]
gemini_predictions = gemini_batch_predict(gemini_examples)

gemini_result = evaluate_predictions(
    gemini_examples,
    gemini_predictions,
    "Gemini",
    GEMINI_MODEL,
)

print_result(gemini_result)


# ============================================================
# OPTIONAL MBT-STYLE SUBCLASS EXPORT
# ============================================================

def mbt_subclass_mapper(ex: TrainingExample) -> str:
    """
    First-pass MBT-style subclass mapping.

    This is not ground truth.
    It is a deterministic taxonomy projection from SNLI binary labels.
    """
    if ex.contradiction_type == "none":
        return "none"

    text = f" {ex.premise} {ex.query} ".lower()

    negation_words = [" not ", " never ", " no ", " nobody ", " nothing ", "n't"]
    if any(w in text for w in negation_words):
        return "direct_negation"

    single_attribute_pairs = [
        ("sitting", "standing"),
        ("standing", "sitting"),
        ("boy", "girl"),
        ("girl", "boy"),
        ("man", "woman"),
        ("woman", "man"),
        ("gray", "black"),
        ("grey", "black"),
        ("black", "gray"),
        ("black", "grey"),
    ]

    for a, b in single_attribute_pairs:
        if a in text and b in text:
            return "property_mismatch"

    color_words = ["black", "white", "blue", "red", "green", "gray", "grey", "yellow", "pink", "brown"]
    if sum(c in text for c in color_words) >= 2:
        return "property_mismatch"

    number_words = ["one", "two", "three", "four", "five", "six", "half"]
    if sum(n in text for n in number_words) >= 2:
        return "property_mismatch"

    return "broad_semantic_contradiction"


snli_mbt_subclass_examples = []

for ex in snli_examples:
    subclass = mbt_subclass_mapper(ex)

    snli_mbt_subclass_examples.append(
        TrainingExample(
            id=ex.id,
            premise=ex.premise,
            query=ex.query,
            contradiction_type=subclass,
            explanation=f"SNLI binary label projected to MBT-style subclass: {subclass}",
            human_patch="",
            repair_action="",
            domain="snli_mbt_subclass_projection",
            difficulty_level=ex.difficulty_level,
        )
    )

print("\nMBT subclass projection counts:")
print(dict(Counter(ex.contradiction_type for ex in snli_mbt_subclass_examples)))


# ============================================================
# SAVE OUTPUTS
# ============================================================

full_report = {
    "config": {
        "snli_split": SNLI_SPLIT,
        "snli_limit": SNLI_LIMIT,
        "snli_seed": SNLI_SEED,
        "gemini_model": GEMINI_MODEL,
        "gemini_limit": GEMINI_LIMIT,
        "gemini_batch_size": GEMINI_BATCH_SIZE,
    },
    "dataset": {
        "name": "SNLI",
        "mapping": {
            "SNLI contradiction": "broad_semantic_contradiction",
            "SNLI entailment": "none",
            "SNLI neutral": "none",
        },
        "label_counts": dict(Counter(ex.contradiction_type for ex in snli_examples)),
    },
    "baseline_results": [asdict(r) for r in baseline_results],
    "gemini_result": asdict(gemini_result),
}

with open(f"{OUTPUT_PREFIX}_full_report.json", "w", encoding="utf-8") as f:
    json.dump(full_report, f, indent=2, ensure_ascii=False)

with open(f"{OUTPUT_PREFIX}_snli_examples.json", "w", encoding="utf-8") as f:
    json.dump([asdict(ex) for ex in snli_examples], f, indent=2, ensure_ascii=False)

with open(f"{OUTPUT_PREFIX}_mbt_subclass_projection.json", "w", encoding="utf-8") as f:
    json.dump([asdict(ex) for ex in snli_mbt_subclass_examples], f, indent=2, ensure_ascii=False)

with open(f"{OUTPUT_PREFIX}_gemini_errors.json", "w", encoding="utf-8") as f:
    json.dump(gemini_result.errors, f, indent=2, ensure_ascii=False)

print("\n" + "=" * 80)
print("DONE")
print("=" * 80)
print(f"Saved: {OUTPUT_PREFIX}_full_report.json")
print(f"Saved: {OUTPUT_PREFIX}_snli_examples.json")
print(f"Saved: {OUTPUT_PREFIX}_mbt_subclass_projection.json")
print(f"Saved: {OUTPUT_PREFIX}_gemini_errors.json")
print("\nCore result:")
print(f"Gemini accuracy: {gemini_result.accuracy:.4f}")
print(f"Gemini macro-F1: {gemini_result.macro_f1:.4f}")
print(f"Gemini examples: {gemini_result.examples_tested}")
print(f"Gemini errors:   {len(gemini_result.errors)}")
````

````python
# ============================================================
# NO-API MODEL BENCHMARK — LOCAL HUGGING FACE NLI MODEL
# ============================================================
# No sign-up.
# No API key.
# No Groq / Gemini / OpenRouter.
# Uses existing snli_examples if present.
# If snli_examples is missing, loads SNLI itself.
# ============================================================

import sys, subprocess, importlib.util, json, time
from dataclasses import dataclass, asdict
from collections import Counter, defaultdict

def ensure_package(package_name, import_name=None):
    import_name = import_name or package_name
    if importlib.util.find_spec(import_name) is None:
        subprocess.check_call([sys.executable, "-m", "pip", "install", "-q", package_name])

ensure_package("datasets")
ensure_package("transformers")
ensure_package("torch")

import torch
from datasets import load_dataset
from transformers import AutoTokenizer, AutoModelForSequenceClassification


@dataclass
class TrainingExample:
    id: str
    premise: str
    query: str
    contradiction_type: str
    explanation: str = ""
    human_patch: str = ""
    repair_action: str = ""
    domain: str = "snli_binary"
    difficulty_level: int = 3


def load_snli_binary(limit=300, split="validation", seed=42):
    ds = load_dataset("stanfordnlp/snli", split=split)
    ds = ds.shuffle(seed=seed)

    examples = []

    for row in ds:
        label = row["label"]
        if label == -1:
            continue

        contradiction_type = "broad_semantic_contradiction" if label == 2 else "none"

        examples.append(
            TrainingExample(
                id=f"snli_{split}_{len(examples):06d}",
                premise=row["premise"],
                query=row["hypothesis"],
                contradiction_type=contradiction_type,
                explanation=f"SNLI label={label} mapped to {contradiction_type}",
            )
        )

        if len(examples) >= limit:
            break

    return examples


# Use existing examples if they exist; otherwise rebuild.
try:
    snli_examples
    print("Using existing snli_examples:", len(snli_examples))
except NameError:
    snli_examples = load_snli_binary(limit=300, split="validation", seed=42)
    print("Loaded new snli_examples:", len(snli_examples))

print("Label counts:", dict(Counter(ex.contradiction_type for ex in snli_examples)))


def evaluate_predictions(examples, predictions, system_name, version):
    labels = sorted(set([ex.contradiction_type for ex in examples] + predictions))
    matrix = defaultdict(Counter)
    errors = []

    for ex, pred in zip(examples, predictions):
        actual = ex.contradiction_type
        matrix[actual][pred] += 1

        if actual != pred:
            errors.append({
                "id": ex.id,
                "actual": actual,
                "predicted": pred,
                "premise": ex.premise,
                "query": ex.query,
            })

    correct = sum(1 for ex, pred in zip(examples, predictions) if ex.contradiction_type == pred)
    accuracy = correct / len(examples) if examples else 0.0

    precision_by_type = {}
    recall_by_type = {}
    f1_by_type = {}

    for label in labels:
        tp = matrix[label][label]
        predicted_as_label = sum(matrix[actual][label] for actual in matrix)
        actual_label = sum(matrix[label].values())

        precision = tp / predicted_as_label if predicted_as_label else 0.0
        recall = tp / actual_label if actual_label else 0.0
        f1 = 0.0 if precision + recall == 0 else 2 * precision * recall / (precision + recall)

        precision_by_type[label] = precision
        recall_by_type[label] = recall
        f1_by_type[label] = f1

    macro_f1 = sum(f1_by_type.values()) / len(f1_by_type) if f1_by_type else 0.0

    return {
        "system_name": system_name,
        "version": version,
        "accuracy": accuracy,
        "macro_f1": macro_f1,
        "precision_by_type": precision_by_type,
        "recall_by_type": recall_by_type,
        "f1_by_type": f1_by_type,
        "confusion_matrix": {actual: dict(preds) for actual, preds in matrix.items()},
        "examples_tested": len(examples),
        "failed_examples": sum(1 for p in predictions if p == "unknown"),
        "errors": errors,
    }


# Smaller/faster than huge BART/RoBERTa-large.
# Good enough for a practical no-API benchmark.
MODEL_NAME = "typeform/distilbert-base-uncased-mnli"

print("\nLoading model:", MODEL_NAME)

tokenizer = AutoTokenizer.from_pretrained(MODEL_NAME)
model = AutoModelForSequenceClassification.from_pretrained(MODEL_NAME)

device = "cuda" if torch.cuda.is_available() else "cpu"
model.to(device)
model.eval()

print("Device:", device)
print("Model labels:", model.config.id2label)


def local_mnli_predict_batch(examples, batch_size=16):
    predictions = []

    # Try to infer which output index means contradiction.
    id2label = {int(k): v.lower() for k, v in model.config.id2label.items()}
    contradiction_ids = [i for i, label in id2label.items() if "contradiction" in label]

    if not contradiction_ids:
        raise RuntimeError(f"Could not find contradiction label in id2label: {model.config.id2label}")

    contradiction_id = contradiction_ids[0]

    for start in range(0, len(examples), batch_size):
        batch = examples[start:start + batch_size]

        premises = [ex.premise for ex in batch]
        hypotheses = [ex.query for ex in batch]

        encoded = tokenizer(
            premises,
            hypotheses,
            padding=True,
            truncation=True,
            return_tensors="pt",
            max_length=256,
        ).to(device)

        with torch.no_grad():
            logits = model(**encoded).logits
            pred_ids = torch.argmax(logits, dim=-1).cpu().tolist()

        for pred_id in pred_ids:
            if pred_id == contradiction_id:
                predictions.append("broad_semantic_contradiction")
            else:
                predictions.append("none")

        print(f"Processed {min(start + batch_size, len(examples))}/{len(examples)}")

    return predictions


# Run 300 examples locally. No API quota.
start_time = time.time()

local_predictions = local_mnli_predict_batch(snli_examples, batch_size=16)

elapsed = time.time() - start_time

local_report = evaluate_predictions(
    snli_examples,
    local_predictions,
    system_name="LocalHF",
    version=MODEL_NAME,
)

local_report["runtime_seconds"] = elapsed
local_report["device"] = device

print("\n" + "=" * 80)
print("LOCAL HUGGING FACE MODEL RESULT")
print("=" * 80)
print(json.dumps({
    "system_name": local_report["system_name"],
    "version": local_report["version"],
    "accuracy": local_report["accuracy"],
    "macro_f1": local_report["macro_f1"],
    "precision_by_type": local_report["precision_by_type"],
    "recall_by_type": local_report["recall_by_type"],
    "f1_by_type": local_report["f1_by_type"],
    "confusion_matrix": local_report["confusion_matrix"],
    "examples_tested": local_report["examples_tested"],
    "failed_examples": local_report["failed_examples"],
    "runtime_seconds": local_report["runtime_seconds"],
    "device": local_report["device"],
}, indent=2))

print("\nFirst 10 errors:")
for err in local_report["errors"][:10]:
    print("\n--- ERROR ---")
    print("ID:", err["id"])
    print("Actual:", err["actual"])
    print("Predicted:", err["predicted"])
    print("Premise:", err["premise"])
    print("Query:", err["query"])

with open("local_hf_snli_300_report.json", "w", encoding="utf-8") as f:
    json.dump(local_report, f, indent=2, ensure_ascii=False)

print("\nSaved: local_hf_snli_300_report.json")
````

````python
# ============================================================
# NEXT STEP — MULTINLI DOMAIN TRANSFER BENCHMARK
# ============================================================
# No API key.
# No signup.
# Reuses RoBERTa-large-MNLI if already loaded.
# Otherwise loads it.
#
# Tests whether the SNLI result generalises to MultiNLI.
# Saves: local_hf_roberta_large_mnli_multinli_300_report.json
# ============================================================

import sys, subprocess, importlib.util, json, time
from dataclasses import dataclass
from collections import Counter, defaultdict

def ensure_package(package_name, import_name=None):
    import_name = import_name or package_name
    if importlib.util.find_spec(import_name) is None:
        subprocess.check_call([sys.executable, "-m", "pip", "install", "-q", package_name])

ensure_package("datasets")
ensure_package("transformers")
ensure_package("torch")

import torch
from datasets import load_dataset
from transformers import AutoTokenizer, AutoModelForSequenceClassification


@dataclass
class TrainingExample:
    id: str
    premise: str
    query: str
    contradiction_type: str
    explanation: str = ""
    human_patch: str = ""
    repair_action: str = ""
    domain: str = "multinli_binary"
    difficulty_level: int = 3


def load_multinli_binary(limit=300, split="validation_mismatched", seed=42):
    """
    MultiNLI labels:
      0 = entailment
      1 = neutral
      2 = contradiction

    Mapping:
      contradiction -> broad_semantic_contradiction
      entailment/neutral -> none
    """
    ds = load_dataset("nyu-mll/multi_nli", split=split)
    ds = ds.shuffle(seed=seed)

    examples = []

    for row in ds:
        label = row["label"]

        contradiction_type = "broad_semantic_contradiction" if label == 2 else "none"

        examples.append(
            TrainingExample(
                id=f"multinli_{split}_{len(examples):06d}",
                premise=row["premise"],
                query=row["hypothesis"],
                contradiction_type=contradiction_type,
                explanation=f"MultiNLI label={label} mapped to {contradiction_type}",
                domain=f"multinli_{split}",
                difficulty_level=3,
            )
        )

        if len(examples) >= limit:
            break

    return examples


def evaluate_predictions(examples, predictions, system_name, version):
    labels = sorted(set([ex.contradiction_type for ex in examples] + predictions))
    matrix = defaultdict(Counter)
    errors = []

    for ex, pred in zip(examples, predictions):
        actual = ex.contradiction_type
        matrix[actual][pred] += 1

        if actual != pred:
            errors.append({
                "id": ex.id,
                "actual": actual,
                "predicted": pred,
                "premise": ex.premise,
                "query": ex.query,
            })

    correct = sum(
        1 for ex, pred in zip(examples, predictions)
        if ex.contradiction_type == pred
    )

    accuracy = correct / len(examples) if examples else 0.0

    precision_by_type = {}
    recall_by_type = {}
    f1_by_type = {}

    for label in labels:
        tp = matrix[label][label]
        predicted_as_label = sum(matrix[actual][label] for actual in matrix)
        actual_label = sum(matrix[label].values())

        precision = tp / predicted_as_label if predicted_as_label else 0.0
        recall = tp / actual_label if actual_label else 0.0
        f1 = 0.0 if precision + recall == 0 else 2 * precision * recall / (precision + recall)

        precision_by_type[label] = precision
        recall_by_type[label] = recall
        f1_by_type[label] = f1

    macro_f1 = sum(f1_by_type.values()) / len(f1_by_type) if f1_by_type else 0.0

    return {
        "system_name": system_name,
        "version": version,
        "accuracy": accuracy,
        "macro_f1": macro_f1,
        "precision_by_type": precision_by_type,
        "recall_by_type": recall_by_type,
        "f1_by_type": f1_by_type,
        "confusion_matrix": {actual: dict(preds) for actual, preds in matrix.items()},
        "examples_tested": len(examples),
        "failed_examples": sum(1 for p in predictions if p == "unknown"),
        "errors": errors,
    }


MODEL_NAME = "FacebookAI/roberta-large-mnli"

# Reuse model2/tokenizer2 if they already exist from the SNLI run.
try:
    tokenizer2
    model2
    device2
    print("Reusing existing RoBERTa model.")
except NameError:
    print("Loading model:", MODEL_NAME)
    tokenizer2 = AutoTokenizer.from_pretrained(MODEL_NAME)
    model2 = AutoModelForSequenceClassification.from_pretrained(MODEL_NAME)
    device2 = "cuda" if torch.cuda.is_available() else "cpu"
    model2.to(device2)
    model2.eval()

print("Device:", device2)
print("Model labels:", model2.config.id2label)


def predict_roberta_binary(examples, batch_size=8):
    predictions = []

    id2label = {int(k): v.lower() for k, v in model2.config.id2label.items()}
    contradiction_ids = [
        i for i, label in id2label.items()
        if "contradiction" in label
    ]

    if not contradiction_ids:
        raise RuntimeError(f"Could not find contradiction label in {model2.config.id2label}")

    contradiction_id = contradiction_ids[0]

    for start in range(0, len(examples), batch_size):
        batch = examples[start:start + batch_size]

        premises = [ex.premise for ex in batch]
        hypotheses = [ex.query for ex in batch]

        encoded = tokenizer2(
            premises,
            hypotheses,
            padding=True,
            truncation=True,
            return_tensors="pt",
            max_length=256,
        ).to(device2)

        with torch.no_grad():
            logits = model2(**encoded).logits
            pred_ids = torch.argmax(logits, dim=-1).cpu().tolist()

        for pred_id in pred_ids:
            if pred_id == contradiction_id:
                predictions.append("broad_semantic_contradiction")
            else:
                predictions.append("none")

        print(f"Processed {min(start + batch_size, len(examples))}/{len(examples)}")

    return predictions


multinli_examples = load_multinli_binary(
    limit=300,
    split="validation_mismatched",
    seed=42,
)

print("\nLoaded MultiNLI examples:", len(multinli_examples))
print("Label counts:", dict(Counter(ex.contradiction_type for ex in multinli_examples)))

start_time = time.time()

multinli_predictions = predict_roberta_binary(multinli_examples, batch_size=8)

elapsed = time.time() - start_time

multinli_report = evaluate_predictions(
    multinli_examples,
    multinli_predictions,
    system_name="LocalHF",
    version=MODEL_NAME + " on MultiNLI validation_mismatched",
)

multinli_report["runtime_seconds"] = elapsed
multinli_report["device"] = device2
multinli_report["dataset"] = "MultiNLI validation_mismatched"
multinli_report["mapping"] = {
    "contradiction": "broad_semantic_contradiction",
    "entailment": "none",
    "neutral": "none",
}

print("\n" + "=" * 80)
print("MULTINLI DOMAIN TRANSFER RESULT")
print("=" * 80)

print(json.dumps({
    "system_name": multinli_report["system_name"],
    "version": multinli_report["version"],
    "dataset": multinli_report["dataset"],
    "accuracy": multinli_report["accuracy"],
    "macro_f1": multinli_report["macro_f1"],
    "precision_by_type": multinli_report["precision_by_type"],
    "recall_by_type": multinli_report["recall_by_type"],
    "f1_by_type": multinli_report["f1_by_type"],
    "confusion_matrix": multinli_report["confusion_matrix"],
    "examples_tested": multinli_report["examples_tested"],
    "failed_examples": multinli_report["failed_examples"],
    "runtime_seconds": multinli_report["runtime_seconds"],
    "device": multinli_report["device"],
}, indent=2))

print("\nFirst 10 errors:")
for err in multinli_report["errors"][:10]:
    print("\n--- ERROR ---")
    print("ID:", err["id"])
    print("Actual:", err["actual"])
    print("Predicted:", err["predicted"])
    print("Premise:", err["premise"])
    print("Query:", err["query"])

with open("local_hf_roberta_large_mnli_multinli_300_report.json", "w", encoding="utf-8") as f:
    json.dump(multinli_report, f, indent=2, ensure_ascii=False)

print("\nSaved: local_hf_roberta_large_mnli_multinli_300_report.json")
````

````python
# ============================================================
# HARD NEGATIVES TEST
# ============================================================
# Goal:
# Build a subset of examples labelled "none" but containing tempting cues:
# no/not/never, attitude words, uncertain inference, extra detail, role/detail shifts.
#
# Tests whether the model over-predicts contradiction on difficult non-contradictions.
# Uses existing:
#   snli_examples
#   multinli_examples
#   multinli_matched_examples
#   predict_roberta_binary
#   evaluate_predictions
# ============================================================

import json
from collections import Counter

def hard_negative_score(ex):
    text = f"{ex.premise} {ex.query}".lower()

    cues = {
        "negation": [" not ", "n't", " no ", " never ", " nobody ", " nothing "],
        "mental_state": ["think", "believe", "fear", "hate", "love", "want", "hope", "sure"],
        "uncertainty": ["maybe", "might", "could", "probably", "presumed", "seems"],
        "extra_specificity": ["tournament", "sidewalk", "home", "doctor", "outside", "inside"],
        "attitude": ["despised", "disregarding", "merciful", "proud", "loud"],
        "temporal": ["before", "after", "prior", "later", "century", "year"],
    }

    score = 0
    hit_types = []

    for cue_type, words in cues.items():
        if any(w in text for w in words):
            score += 1
            hit_types.append(cue_type)

    return score, hit_types


def collect_hard_negatives(datasets, limit=150):
    candidates = []

    for dataset_name, examples in datasets:
        for ex in examples:
            if ex.contradiction_type != "none":
                continue

            score, hit_types = hard_negative_score(ex)

            if score > 0:
                candidates.append({
                    "dataset": dataset_name,
                    "score": score,
                    "hit_types": hit_types,
                    "example": ex,
                })

    candidates = sorted(candidates, key=lambda x: x["score"], reverse=True)
    return candidates[:limit]


datasets_for_hard_negatives = [
    ("SNLI", snli_examples),
    ("MultiNLI mismatched", multinli_examples),
    ("MultiNLI matched", multinli_matched_examples),
]

hard_negative_items = collect_hard_negatives(
    datasets_for_hard_negatives,
    limit=150,
)

hard_negative_examples = [item["example"] for item in hard_negative_items]

print("Hard negative examples:", len(hard_negative_examples))
print("Source counts:", dict(Counter(item["dataset"] for item in hard_negative_items)))
print("Cue counts:", dict(Counter(cue for item in hard_negative_items for cue in item["hit_types"])))

print("\nFirst 10 hard negatives:")
for item in hard_negative_items[:10]:
    ex = item["example"]
    print("\nDataset:", item["dataset"])
    print("Score:", item["score"])
    print("Cues:", item["hit_types"])
    print("Premise:", ex.premise)
    print("Query:", ex.query)

hard_negative_predictions = predict_roberta_binary(
    hard_negative_examples,
    batch_size=8,
)

hard_negative_report = evaluate_predictions(
    hard_negative_examples,
    hard_negative_predictions,
    system_name="LocalHF",
    version=MODEL_NAME + " hard negatives",
)

hard_negative_report["dataset"] = "Hard negatives pooled from SNLI + MultiNLI"
hard_negative_report["source_counts"] = dict(Counter(item["dataset"] for item in hard_negative_items))
hard_negative_report["cue_counts"] = dict(Counter(cue for item in hard_negative_items for cue in item["hit_types"]))

print("\n" + "=" * 80)
print("HARD NEGATIVES RESULT")
print("=" * 80)

print(json.dumps({
    "dataset": hard_negative_report["dataset"],
    "n": hard_negative_report["examples_tested"],
    "accuracy": hard_negative_report["accuracy"],
    "macro_f1": hard_negative_report["macro_f1"],
    "precision_by_type": hard_negative_report["precision_by_type"],
    "recall_by_type": hard_negative_report["recall_by_type"],
    "f1_by_type": hard_negative_report["f1_by_type"],
    "confusion_matrix": hard_negative_report["confusion_matrix"],
    "failed_examples": hard_negative_report["failed_examples"],
    "source_counts": hard_negative_report["source_counts"],
    "cue_counts": hard_negative_report["cue_counts"],
}, indent=2))

print("\nFalse positives on hard negatives:")
for err in hard_negative_report["errors"][:20]:
    print("\n--- FALSE POSITIVE ---")
    print("ID:", err["id"])
    print("Actual:", err["actual"])
    print("Predicted:", err["predicted"])
    print("Premise:", err["premise"])
    print("Query:", err["query"])

with open("hard_negative_roberta_large_mnli_report.json", "w", encoding="utf-8") as f:
    json.dump(hard_negative_report, f, indent=2, ensure_ascii=False)

with open("hard_negative_examples.json", "w", encoding="utf-8") as f:
    json.dump([
        {
            "dataset": item["dataset"],
            "score": item["score"],
            "hit_types": item["hit_types"],
            "example": item["example"].__dict__,
        }
        for item in hard_negative_items
    ], f, indent=2, ensure_ascii=False)

print("\nSaved: hard_negative_roberta_large_mnli_report.json")
print("Saved: hard_negative_examples.json")
````

````python
# ============================================================
# HARD POSITIVES TEST
# ============================================================
# Goal:
# Build contradiction examples with weak/no explicit negation cues.
# These test semantic contradiction detection rather than easy negation.
#
# Uses existing:
#   snli_examples
#   multinli_examples
#   multinli_matched_examples
#   predict_roberta_binary
#   evaluate_predictions
# ============================================================

import json
from collections import Counter

def hard_positive_score(ex):
    text = f"{ex.premise} {ex.query}".lower()

    negation_cues = [" not ", "n't", " no ", " never ", " nobody ", " nothing ", "without"]

    semantic_cues = {
        "scene_shift": ["inside", "outside", "store", "field", "home", "street", "beach", "water"],
        "action_shift": ["running", "sleeping", "standing", "sitting", "walking", "working", "playing", "riding", "knitting"],
        "role_entity_shift": ["man", "woman", "boy", "girl", "children", "workers", "people"],
        "quantity_shift": ["one", "two", "three", "four", "five", "six", "half", "majority"],
        "attribute_shift": ["black", "white", "blue", "gray", "grey", "red", "green", "large", "small"],
        "conditional_swap": ["if", "when", "stands still", "runs away"],
        "document_reference": ["figure", "table", "sku", "weekly", "monthly"],
    }

    has_negation = any(cue in text for cue in negation_cues)

    score = 0
    hit_types = []

    for cue_type, words in semantic_cues.items():
        hits = sum(w in text for w in words)
        if hits >= 2:
            score += 1
            hit_types.append(cue_type)

    # Prefer hard positives without explicit negation.
    if not has_negation:
        score += 1
        hit_types.append("no_explicit_negation")

    return score, hit_types, has_negation


def collect_hard_positives(datasets, limit=150):
    candidates = []

    for dataset_name, examples in datasets:
        for ex in examples:
            if ex.contradiction_type != "broad_semantic_contradiction":
                continue

            score, hit_types, has_negation = hard_positive_score(ex)

            if score > 0:
                candidates.append({
                    "dataset": dataset_name,
                    "score": score,
                    "hit_types": hit_types,
                    "has_negation": has_negation,
                    "example": ex,
                })

    candidates = sorted(candidates, key=lambda x: x["score"], reverse=True)
    return candidates[:limit]


datasets_for_hard_positives = [
    ("SNLI", snli_examples),
    ("MultiNLI mismatched", multinli_examples),
    ("MultiNLI matched", multinli_matched_examples),
]

hard_positive_items = collect_hard_positives(
    datasets_for_hard_positives,
    limit=150,
)

hard_positive_examples = [item["example"] for item in hard_positive_items]

print("Hard positive examples:", len(hard_positive_examples))
print("Source counts:", dict(Counter(item["dataset"] for item in hard_positive_items)))
print("Cue counts:", dict(Counter(cue for item in hard_positive_items for cue in item["hit_types"])))
print("Negation counts:", dict(Counter(item["has_negation"] for item in hard_positive_items)))

print("\nFirst 10 hard positives:")
for item in hard_positive_items[:10]:
    ex = item["example"]
    print("\nDataset:", item["dataset"])
    print("Score:", item["score"])
    print("Cues:", item["hit_types"])
    print("Has negation:", item["has_negation"])
    print("Premise:", ex.premise)
    print("Query:", ex.query)

hard_positive_predictions = predict_roberta_binary(
    hard_positive_examples,
    batch_size=8,
)

hard_positive_report = evaluate_predictions(
    hard_positive_examples,
    hard_positive_predictions,
    system_name="LocalHF",
    version=MODEL_NAME + " hard positives",
)

hard_positive_report["dataset"] = "Hard positives pooled from SNLI + MultiNLI"
hard_positive_report["source_counts"] = dict(Counter(item["dataset"] for item in hard_positive_items))
hard_positive_report["cue_counts"] = dict(Counter(cue for item in hard_positive_items for cue in item["hit_types"]))
hard_positive_report["negation_counts"] = dict(Counter(str(item["has_negation"]) for item in hard_positive_items))

print("\n" + "=" * 80)
print("HARD POSITIVES RESULT")
print("=" * 80)

print(json.dumps({
    "dataset": hard_positive_report["dataset"],
    "n": hard_positive_report["examples_tested"],
    "accuracy": hard_positive_report["accuracy"],
    "macro_f1": hard_positive_report["macro_f1"],
    "precision_by_type": hard_positive_report["precision_by_type"],
    "recall_by_type": hard_positive_report["recall_by_type"],
    "f1_by_type": hard_positive_report["f1_by_type"],
    "confusion_matrix": hard_positive_report["confusion_matrix"],
    "failed_examples": hard_positive_report["failed_examples"],
    "source_counts": hard_positive_report["source_counts"],
    "cue_counts": hard_positive_report["cue_counts"],
    "negation_counts": hard_positive_report["negation_counts"],
}, indent=2))

print("\nFalse negatives on hard positives:")
for err in hard_positive_report["errors"][:20]:
    print("\n--- FALSE NEGATIVE ---")
    print("ID:", err["id"])
    print("Actual:", err["actual"])
    print("Predicted:", err["predicted"])
    print("Premise:", err["premise"])
    print("Query:", err["query"])

with open("hard_positive_roberta_large_mnli_report.json", "w", encoding="utf-8") as f:
    json.dump(hard_positive_report, f, indent=2, ensure_ascii=False)

with open("hard_positive_examples.json", "w", encoding="utf-8") as f:
    json.dump([
        {
            "dataset": item["dataset"],
            "score": item["score"],
            "hit_types": item["hit_types"],
            "has_negation": item["has_negation"],
            "example": item["example"].__dict__,
        }
        for item in hard_positive_items
    ], f, indent=2, ensure_ascii=False)

print("\nSaved: hard_positive_roberta_large_mnli_report.json")
print("Saved: hard_positive_examples.json")
````
